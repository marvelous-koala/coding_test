```
## 베스트 엘범(해쉬)
# 스트리밍 사이트에서 장르 별로 가장 많이 재생된 노래를 두 개씩 모아 베스트 앨범을 출시하려 합니다
def solution(genres, plays):
    answer = []
    d = {e:[] for e in set(genres)}
    for e in zip(genres, plays, range(len(plays))):
        d[e[0]].append([e[1] , e[2]])
    genreSort =sorted(list(d.keys()), key= lambda x: sum( map(lambda y: y[0],d[x])), reverse = True)
    for g in genreSort:
        temp = [e[1] for e in sorted(d[g],key= lambda x: (x[0], -x[1]), reverse = True)]
        answer += temp[:min(len(temp),2)]
    return answer
```

```
## 다리를 지나는 트럭(스택/큐)

def solution(bridge_length, weight, truck_weights):
    wait = truck_weights[::-1]
    on_bridge =[0]*bridge_length
    tick = 0
    while True:
        on_bridge[1:] = on_bridge[:-1]
        on_bridge[0] = 0  
        tick += 1
        if len(wait) == 0 and sum(on_bridge) == 0:
            break
        elif len(wait) > 0 and wait[-1] + sum(on_bridge) <= weight:
            on_bridge[0] = wait.pop()
    return tick
```

```
## 프린터
# 중요도가 높은 문서를 먼저 인쇄하는 프린터를 개발했습니다. 이 새롭게 개발한 프린터는 아래와 같은 방식으로 인쇄 작업을 수행합니다.

import heapq

def solution(operations):
    heap = []
    for opString in operations:
        op, num = opString.split()
        if op == "I":
            heapq.heappush(heap, int(num))
        elif heap:
            if num[0] == '-':
                heapq.heappop(heap)
            else:
                heap.remove(max(heap))
    return [0, 0] if not heap else [max(heap), min(heap)]
```

```
## 소수 찾기 (완전 탐색)
# 한자리 숫자가 적힌 종이 조각이 흩어져있습니다. 흩어진 종이 조각을 붙여 소수를 몇 개 만들 수 있는지 알아내려 합니다.

from itertools import permutations

def solution(numbers):
    sp = [i for i in numbers]
    chek_list = set()
    ans = 0
    for nums in range(1, len(sp) + 1):    
        combi = list(permutations(sp, nums))       
        for k in combi:
            a = ''.join(k)
            chek_list.add(int(a))
    chek_list = [i for i in chek_list if i > 1]
    for item in chek_list:
        ab = True
        for i in range(2, item):
            if (item/i) - (item//i) == 0:
                ab = False
                break
        if ab:
            ans +=1
        
    return ans
```

```
## 그리디 서치 (체육복)
# 점심시간에 도둑이 들어, 일부 학생이 체육복을 도난당했습니다. 다행히 여벌 체육복이 있는 학생이 이들에게 체육복을 빌려주려 합니다.

def solution(n, lost, reserve):
    student = [1] * (n + 2)
    for i in lost:
        student[i] -= 1
    for i in reserve:
        student[i] += 1
    for i in range(1, n+1):
        if student[i] == 0 and student[i-1] ==2:
            student[i-1 : i+1] = [1,1]
        elif student[i] == 0 and student[i +1] ==2:
            student[i : i+2] = [1,1]
    return len([x for x in student if x > 0]) - 2
```

```
## 조이스틱(그리드 서치)
# 조이스틱으로 알파벳 이름을 완성하세요. 맨 처음엔 A로만 이루어져 있습니다.

def solution(name):
    m = [ min(ord(c) - 65, 91-ord(c)) for c in name]       

    answer = 0
    where = 0

    while True:    
        answer += m[where]
        m[where] = 0

        if sum(m) == 0:
            break

        left, right = (1,1)

        while m[where - left] <= 0:
            left += 1
        while m[where + right] <= 0:
            right += 1

        answer += left if left < right else right
        where += -left if left < right else right

    return answer
```

```
## 큰 수 만들기(그리드 서치)
# 어떤 숫자에서 k개의 수를 제거했을 때 얻을 수 있는 가장 큰 숫자를 구하려 합니다.

def solution(number, k):
    cols = []
    for i , num in enumerate(number):
        while len(cols) > 0 and cols[-1] < num and k > 0:
            cols.pop()
            k -= 1
        if k == 0:
            cols += list(number[i:])
            break
        cols.append(num)
    cols = cols[:-k] if k > 0 else cols
    answer = ''.join(cols)
    return answer
```

```
## 구명보트 (그리드 서치)
# 무인도에 갇힌 사람들을 구명보트를 이용하여 구출하려고 합니다. 구명보트는 작아서 한 번에 최대 2명씩 밖에 탈 수 없고, 무게 제한도 있습니다.

def solution(people, limit):
    answer = 0
    people.sort()

    f_cnt =0
    e_cnt =len(people)-1
    while e_cnt - f_cnt >=1:
        if people[f_cnt] +people[e_cnt] <= limit:
            f_cnt +=1
            e_cnt -=1
        else:
            e_cnt -=1
        answer +=1
    if e_cnt == f_cnt:
        answer +=1
    return answer
```

```
## N으로 표현(동적계획법)
# 아래와 같이 5와 사칙연산만으로 12를 표현할 수 있습니다.

def solution(N, number):
    s = [set() for x in range(8)]
    for i, x in enumerate(s, start = 1):
        x.add(int(str(N) * i))
    for i in range(1, len(s)):
        for j in range(i):
            for val1 in s[j]:
                for val2 in s[i -j -1]:
                    s[i].add(val1 + val2)
                    s[i].add(val1 - val2)
                    s[i].add(val2 - val1)
                    s[i].add(val1 * val2)
                    if val2 != 0:
                        s[i].add(val1//val2)
                    if val1 != 0:
                        s[i].add(val2//val1)
        if number in s[i]:
            answer = i + 1
            break
    else:
        answer = -1
    return answer
```

```
## 트라이앵글(동적계획법)
# 위와 같은 삼각형의 꼭대기에서 바닥까지 이어지는 경로 중, 거쳐간 숫자의 합이 가장 큰 경우를 찾아보려고 합니다.

def solution(triangle):
    t_h = len(triangle)
    t_w = [ [0 for _ in range(layer+1)] for layer in range(t_h) ]

    t_w[0][0] = triangle[0][0]
    for layer in range(1, t_h):        
        for room in range(layer+1):
            if room == 0:
                    t_w[layer][room] = triangle[layer][room] + t_w[layer-1][room]
            elif room == layer:
                    t_w[layer][room] = triangle[layer][room] + t_w[layer-1][room-1]
            else: 
                    t_w[layer][room] = triangle[layer][room] + max(t_w[layer-1][room-1], t_w[layer-1][room])

    return max(t_w[t_h-1])
```

```
## 타겟넘버(깊이/너비 우선)
# n개의 음이 아닌 정수가 있습니다. 이 수를 적절히 더하거나 빼서 타겟 넘버를 만들려고 합니다. 

from itertools import product
def solution(numbers, target):
    li = [(x, -x) for x in numbers]
    k = list(map(sum, product(*li)))
    return k.count(target)
```

```
## 네트워크(깊이/너비우선)
# 네트워크란 컴퓨터 상호 간에 정보를 교환할 수 있도록 연결된 형태를 의미합니다.

def dfs(computers, start, explored):
    if start in explored:
        return 0, explored
    explored.add(start)
    neighbors = computers[start]
    for ni in range(len(neighbors)):
        if ni == start or neighbors[ni] == 0: continue
        if ni not in explored:
            dfs(computers, ni, explored)
    return 1, explored

def solution(n, computers):
    answer = 0
    explored = set()
    for start in range(0, n):
        res, explored = dfs(computers, start, explored)
        print('start', start)
        print('explored', explored)
        answer += res
    return answer
```

```
## 여행경로 (깊이/너비우선)
# 주어진 항공권을 모두 이용하여 여행경로를 짜려고 합니다. 항상 "ICN" 공항에서 출발합니다.

def solution(tickets):
    routes = {}
    
    for t in tickets:
        routes[t[0]] = routes.get(t[0], []) + [t[1]]
    for r in routes:
        routes[r].sort(reverse = True)
    stack = ['ICN']
    path = []
    while len(stack) > 0:
        top = stack[-1]
        if top not in routes or len(routes[top]) == 0:
            path.append(stack.pop())
        else:
            stack.append(routes[top].pop())
    
    return path[::-1]
```

```
## 입국심사(이분탐색)
# n명이 입국심사를 위해 줄을 서서 기다리고 있습니다. 각 입국심사대에 있는 심사관마다 심사하는데 걸리는 시간은 다릅니다.

def solution(n, times):
    answer = 0
    # left: 최소시간, right: 최대 시간
    left = 1; right = min(times)*n;
    # 입국심사를 하는데, 걸리는 시간의 총합이 가장 작은 것을 찾는다.
    while left <= right:
        # 중앙값: 심사위원에게 주어진 시간
        mid = (left+right)//2
        # 모든 심사위원이 처리한 사람 수 
        cnt = 0
        for t in times:
            cnt += mid // t
        # 대기 중인 사람보다 더 많이 처리하면 mid를 줄일 수 있다.
            if cnt >= n:
                right = mid-1
                answer = mid
                break
        # 대기 중인 사람보다 더 적게 처리하면 mid가 커져야 한다.
        if cnt < n:
            left =mid+1

    return answer
```

```
## 가장 먼 노드(그래프)
# BFS 방식, n개의 노드가 있는 그래프가 있습니다. 각 노드는 1부터 n까지 번호가 적혀있습니다. 1번 노드에서 가장 멀리 떨어진 노드의 갯수를 구하려고 합니다. 

def solution(n, edge):
    import collections
    answer = [1]+[0]*(n-1)
    vec = [[]for i in range(n)]
    que = collections.deque([1])
    for i in edge:
        vec[i[0]-1].append(i[1])
        vec[i[1]-1].append(i[0])
    while que:
        l = len(que)
        for i in range(l):
            q = que.popleft()
            for e in vec[q-1]:
                if answer[e-1] == 0:
                    answer[e-1] = 1
                    que.append(e)
    return l
```
